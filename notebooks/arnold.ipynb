```python
import matplotlib.pyplot as plt
from matplotlib import image
import numpy as np
```

```python
def arnold(src):
    a = np.array([[2,1],[1,1]])
    M, N = src.shape
    points = np.mgrid[0:N, 0:M].reshape((2, M*N))
    new_points = a.dot(points)%[[M], [N]]
    x, y = new_points.reshape((2, M, N), order='F')
    indices = x + N*y
    return np.take(src, indices, mode='wrap')
```

```python
init = np.tril(np.ones((20,20)))
cur = init
N = 6
fig, axs = plt.subplots(N, N, figsize=(10,10))
for i in range(N):
    for j in range(N):
        axs[i, j].axis('off')
        axs[i, j].imshow(cur)
        cur = arnold(cur)
```
